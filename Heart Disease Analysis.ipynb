# Plot Continuous Variables

```python
df.head()
```

```python
cate_values = []
cont_values = []
for column in df.columns:
    if df[column].nunique() <= 10:
        cate_values.append(column)
    else:
        cont_values.append(column)
```

```python
cate_values
```

```python
cont_values
```

```python
df.hist(cont_values, figsize=(15, 6))
plt.tight_layout()
plt.show()
```
